# 1から6の数字を10個発生させる

```python
import numpy as np
temp = np.random.randint(1, 7, size=[10, 1])

# 表示
temp
```

```python
# ヒストグラム
import seaborn as sns
sns.displot(temp)
```

# 1から6の数字を10個発生させて、5回分の合計値を算出する

```python
import numpy as np
temp = np.random.randint(1, 7, size=[10, 5])
x = temp.sum(axis=1)

# 表示
x
```

```python
# ヒストグラム
import seaborn as sns
sns.displot(temp)
```
