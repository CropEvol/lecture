# 1から6の数字を10個発生させる

```python
import numpy as np
temp = np.random.randint(1, 7, size=[10, 1])
```

```python
# ヒストグラム
import matplotlib.pyplot as plt
plt.hist(temp)
plt.show()
```

# 1から6の数字を10個発生させて、5回分の合計値を算出する

```python
import numpy as np
temp = np.random.randint(1, 7, size=[10, 5])
x2 = temp.sum(axis=1)
```

```python
# ヒストグラム
import matplotlib.pyplot as plt
plt.hist(x2)
plt.show()
```
