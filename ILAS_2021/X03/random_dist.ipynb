# 乱数の生成とその分布

　NumPyライブラリを使うと、任意個数の乱数（ランダムな数値データ）を生成させられます。

## 例1) 一様分布に従って（ランダムに）、0から1の範囲の数値を生成させる

`numpy.random.rand(size)`で、0 <= x < 1 の数値（小数点数）をsizeに指定した個数生成できます。

```python
import numpy as np

# 出力: 1個
x1 = np.random.rand(1)
print(x1)
```

### 練習1）

　`np.random.rand()`に与えている数値を`10`に変更するとどうなるか確かめてみてください。

## 例2) 一様分布に従って（ランダムに）、任意の範囲の"整数"を生成させる

`numpy.random.randint(a, b, size)`で、a <= x < b の整数をsizeに指定した個数生成できます。

```python
import numpy as np

# 出力: 1から6の整数を１個
x2 = np.random.randint(1, 7, size=1)
print(x2)
```

### 練習2）

　`np.random.randint()`の`size`に与えている数値を`10`に変更するとどうなるか確かめてみてください。

## 例3） 発生させた乱数をヒストグラム（頻度分布図）にする

seabornライブラリの`displot`を使うと、簡単にヒストグラムを描くことができます。

```python
import numpy as np

# 出力: 1から6の整数を１0個
x3 = np.random.randint(1, 7, size=10)
print(x3)
```

```python
# ヒストグラム
import seaborn as sns
sns.displot(x3, discrete=True)
```

補足: `discrete=True`は、離散値データのヒストグラムを綺麗に表示するためのオプションです。

### 練習3）
　`np.random.randint()`の`size`に与えている数値を`10`や`1000`、`10000`に変更して、ヒストグラムを描くとどうなるか確かめてみてください。

## 例4） 9個のサイコロを3回振る

　リスト形成で複数の数値`size`に与えると、n行m列の乱数行列を作ることができます。

　別の表現にすると、「n個のサイコロをm回振る」といった乱数を生成できます。


```python
import numpy as np
x4 = np.random.randint(1, 7, size=[15, 20])

# 表示
x4
```

### 練習4）

　コードを変更して、「15行20列」の乱数行列を作ってください。

## 例5） 各回のサイコロの合計値をヒストグラムにする

　n行m列（「n個のサイコロをm回振る」）の乱数行列に、`.sum(axis=0)`を付けることで、各列の合計値を求めることができます。

　その合計値をヒストグラムにすると、どうなるでしょうか？

```python
import numpy as np
t = np.random.randint(1, 7, size=[9, 3])
x5 = t.sum(axis=0)

# 表示
x5
```

```python
# ヒストグラム
import seaborn as sns
sns.displot(x5, discrete=True)
```

### 練習5）
 9行1000列（「9個のサイコロを1000回振る」）をおこない、乱数行列の各列の合計値を求めて、ヒストグラムにしてください。
