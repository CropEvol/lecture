## 課題2

　データフレーム`df`から、`N_ALT`の列データを取り出してください。

### 解答例

　`データフレーム.loc[行名, 列名]`の形で行名と列名を指定すると、データフレーム上の任意のデータにアクセスできます。今回のデータの場合、`df[2, "POS"]`とすると、行名`2`と列名`POS`で指定されるデータにアクセスできます。

<img src="https://lh3.googleusercontent.com/pw/ACtC-3dTmqWh96aiKR7r6N9bCpmm8v727Axn8Lr-bIv4OGgKOy3muakVuUe5H_VQMJdyfVyiq-qj8EMj8VwAP7iUqoXkji6-8hKKeLUzYHnsqwEmKgZeBwHysYqIGofizfXjOD_n7SoD8wtzpB2YWe6MDVQl=w1335-h670-no?authuser=0" alt="pandas_dataframe" height="200px">

　列名`N_ALT`のすべての行データにアクセスしたい場合、行名の部分にコロン記号(`:`)を書きます。このコロン記号(`:`)は、"最初の行から最後の行まで(すなわち、すべての行）"として解釈されます。

````
df.loc[:, "N_ALT"]
````

　1行のデータにアクセスしたい場合は、下記のように簡単に記述することも可能です。

```python

データフレーム.列名
```

````
# 以下のどちらかのコードでも可
df["N_ALT"]
df.N_ALT
````

## 課題3

　データフレーム`df`から、`N_ALT`の値が**0以外**の行データを取り出してください。

### 解答例

　条件に合う行データを取り出す方法は、`データフレーム[ 条件式 ]`です。

　条件式の部分に「N_ALTの値が0以外」を表すコードを書きます。
```python
# 条件式の部分
# 「!=」は "等しくない" を表します
df["N_ALT"] != 0
```

````
df[df["N_ALT"] != 0]
````

## 課題4

<small>※ これは取り組み必須課題ではありません。</small>

　2つの列データ`N_REF`と`N_ALT`について下記の計算をおこなってください。また。その計算結果を新しい列データ`Freq_REF`として、データフレーム`df`に追加してください。

```python
# リファレンスと同じ塩基をもつリードの割合
N_REF/(N_REF + N_ALT)
```



### 解答例

　列同士の計算は下記のように簡単におこなえます。
```python
# 書き方
列Aデータ + 列Bデータ
# 例: 足し算
df["N_REF"] + df["N_ALT"]
```

　新しい列を追加するには、次のように書きます。
```python
# 書き方
データフレーム[新しい列名] = 追加したいデータ
# 例: 足し算の結果を新しい列tashizanに追加する
df["tashizan"] = df["N_REF"] + df["N_ALT"]
```

````
df["Freq_REF"] = df["N_REF"]/(df["N_REF"] + df["N_ALT"])

# 確認
df
````

## 課題5
<small>※ これは取り組み必須課題ではありません。</small>

　課題4で追加した列データ`Freq_REF`の情報を使って、「`Freq_REF`が0.1以上かつ0.9以下の行データ」を取り出してください。


### 解答例

　条件式が2つ以上の場合、各条件に丸括弧を使います。
```
# AND条件: 条件A かつ 条件B
データフレーム[ (条件式A) & (条件式B) ]

# OR条件: 条件A または 条件B
データフレーム[ (条件式A) | (条件式B) ]
```

````
df[ (df["Freq_REF"] >= 0.1) & (df["Freq_REF"] <= 0.9) ]
````
