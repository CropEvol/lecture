#### 解答例（正解例）

　Pythonでのファイル処理の基本は、「上の行から1行ずつ順番に処理する」です。1行のデータを処理するときに、1行のデータを分割して、必要な情報のみを取り出せるようにします。そして、必要な情報を使って、データ処理をおこないます。

　今回の課題では、下のようなコードを順番に追加していくと、プログラムを完成させられます。

1. ファイルを開く: `open()`・ファイルを閉じる: `close()`
1. 1行ずつデータを取り出す: `for`構文
1. 1行を分割して、リストにする: `split()`
1. リストから要素（DNA配列本体）を取り出す: `リスト変数[インデックス]`
1. 取り出したDNA配列本体に対して、`count()`や`len()`を実行する
1. `count()`や`len()`の実行結果を使って計算する

<img src="https://lh3.googleusercontent.com/pw/ACtC-3fWrQ0GQAuNX6_O-d34BU_ypbleVReqV1LiarxUOkZgVnSPEdK74fNkxXhMJgM5SjZhjhslghw2h-pu18riKXJIVvVfOVIF-1Hoinm_iN2HQGi8M3BlTY0HDlqWXdA5TdMZ1RM_70YHdDCyMTQYMMtb=w1885-h874-no?authuser=0" alt="file_processing" height="250px" align="middle">



````
filename = "ten_sequence.txt"

# ファイルを開く
f = open(filename)

# 1行ずつデータを取り出し、処理する
for line in f:
  # 1行を分割し、リストデータにする
  line = line.split()

  # リストデータから要素を取り出す
  name      = line[0]  # DNA配列の名前
  sequence  = line[1]  # DNA配列の本体

  # Nの割合を調べる
  freq_N = sequence.count("N") / len(sequence)

  # 表示する
  print(name, freq_N)

# ファイルを閉じる
f.close()
````

#### 間違い例

　下記のコードでは、DNA配列名とDNA配列本体の両方を含んだ1行のデータに対して`len()`や`count()`を使用しています。

````
filename = "ten_sequence.txt"

# ファイルを開く
f = open(filename)

# 1行ずつデータを取り出し、処理する
for line in f:

  length   = len(line)         # 文字列の長さ（問題点1）
  count_N = line.count("N")  # 文字列中のNの数（問題点2）

  # Nの割合を調べる
  freq_N = count_N / length

  # 表示する
  print(freq_N)

# ファイルを閉じる
f.close()
````

問題点

1. `len()`を実行すると、DNA配列名（sequence_Xの部分）や半角スペース、改行文字を含めた文字列の長さ（文字数）を調べてしまっています。
2. 今回のサンプルデータでは、各DNA配列の名前に大文字の「N」が含まれていないため、1行のデータ全体に対して`count()`を実行しても誤った結果にはなりません。しかし、DNA配列名にNが含まれている場合、問題が生じてしまいます。

改善方法
- `split()`で1行のデータを分割して、DNA配列本体のみを取り出し、DNA配列本体のみに対して`len()`や`count()`を実行する



#### 正解ではあるが、改善したほうが良い例

　下のコードは、実行すると正しい出力結果が得られます。しかし、正しい結果は得られていますが、改善したほうが良い箇所があります。

````
filename = "ten_sequence.txt"

# ファイルを開く
f = open(filename)

# 1行ずつデータを取り出し、処理する
for line in f:
  dna = line[11:91]          # DNA配列本体（改善すべき箇所）
  length   = len(dna)        # DNA配列本体の長さ
  count_N = dna.count("N") # DNA配列本体中のNの数

  # Nの割合を調べる
  freq_N = count_N / length
  
  # 表示する
  print(freq_N)

# ファイルを閉じる
f.close()
````

改善方法

### おまけ

　上の課題は、『個々のDNA配列の「未定義塩基（N）」の割合を調べる』でした。

　ここでは、全部DNA配列をまとめたときの「Nの割合」を調べる方法を考えてみましょう。

2通りの方法があります。
1. 各配列を処理する過程でDNA配列の「長さ」と「Nの個数」の積算し、最後に積算値で「Nの割合」を求める。
1. 各DNA配列を結合して、1本の大きなDNA配列を作り、その配列中の「Nの割合」を求める。

#### 方法1: 積算をおこなう方法


````
filename = "ten_sequence.txt"

# ファイルを開く
f = open(filename)

# 積算用の変数
total_L = 0 # DNA配列の長さの積算用変数（初期値:0）
total_N = 0 # Nの個数の積算用変数 （初期値:0）

# 1行ずつデータを取り出し、処理する
for line in f:
  # 1行を分割し、リストデータにする
  line = line.split()

  # リストデータから要素を取り出す
  sequence  = line[1]  # DNA配列の本体

  # 積算（数値計算をして、元の変数に再代入する）
  total_L = total_L + len(sequence)
  total_N = total_N + sequence.count("N")

# ファイルを閉じる
f.close()

print(total_N) # DNA配列の長さの積算
print(total_L) # Nの個数の積算
print(total_N/total_L) # Nの割合
````

#### 方法2: 一つの大きなDNA配列を作る方法

````
filename = "ten_sequence.txt"

# ファイルを開く
f = open(filename)

dna = "" # DNA配列結合用の変数（初期値：空文字列）

# 1行ずつデータを取り出し、処理する
for line in f:
  # 1行を分割し、リストデータにする
  line = line.split()

  # DNA配列を足し合わせる
  # （文字列結合をして、元の変数に再代入する）
  dna = dna + line[1]

# ファイルを閉じる
f.close()

print(len(dna))        # 大きなDNA配列の長さ
print(dna.count("N")) # 大きなDNA配列中のNの個数
print(dna.count("N")/len(dna)) # Nの割合
````
