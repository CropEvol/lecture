## 課題1: プログラムを実行する

　次のコードセルのプログラムを実行してください。

````
# DNA配列中の"A"の個数を調べるプログラム
dna = "GGCTTATTAGCTGAA" 
cnt = dna.count("A")
print(cnt)
````

## 課題2: プログラムを一部書き換えて、実行する

　課題1のプログラムの`cnt = dna.count("A")`を、次のように変更して実行してください。
 
```python
cnt = dna.count("GCT")
```

````
# 下記のプログラムを書き換えて実行する
dna = "GGCTTATTAGCTGAA" 
cnt = dna.count("A")
print(cnt)
````

## 課題3: 共有リンクを発行する
　次の手順で、共有リンクを発行してください。

1. ノートブックを保存する
  - 「ファイル」 > 「保存」または「ドライブにコピーを保存」
1. 共有リンクを発行する  
  - 右上の「共有」ボタンをクリックする
  - 「リンクを知っている全員に変更」をクリックする
  - 権限が「閲覧者」になっているか確認する
  - 共有リンクをコピーする
