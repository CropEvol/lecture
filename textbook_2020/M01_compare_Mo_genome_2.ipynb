　実験では、下の写真のように培地上で培養でき、条件が揃うといもち病菌は胞子を作ります。そして、胞子を植物に接種すると、そのいもち病菌が植物に感染するかどうかを調べられます。

　解析では、**感染する植物の異なる2つのいもち病菌**のデータを扱います。

- イネいもち病菌：TH3O  
- アワいもち病菌：Sv11  
<small>※ TH3OやSv11というのは、いもち病菌系統に名付けられた名前で、亜種や変種レベルの正式な学術名ではありません。</small>




この2種類は、同じいもち病菌ですが、感染できる植物（ホスト）が異なります。
- **イネいもち病菌** → **イネに感染できる**　/ アワに感染できない
- **アワいもち病菌** → イネに感染できない　/ **アワに感染できる**

　感染できるホストの違い（**宿主特異性; Host specificity**）は、どのような機構で生じているのでしょうか？

　宿主特異性は、「植物（ホスト）側の遺伝子」と「いもち病菌側の遺伝子」の相互作用で決まっていると考えられています。

- イネ: いもち病菌の遺伝子Aから作られるタンパク質Aを認識する
  - イネいもち病菌: 遺伝子Aをもたない → イネに認識されない（感染できる）
  - アワいもち病菌: 遺伝子Aをもつ → イネに認識される（感染できない）
- アワ: いもち病菌の遺伝子Bから作られるタンパク質Bを認識する
  - イネいもち病菌: 遺伝子Bをもつ → アワに認識される（感染できない）
  - アワいもち病菌: 遺伝子Bをもたない → アワに認識されない（感染できる）


　遺伝子A（アワいもち病菌にはあり、イネいもち病菌には存在しない遺伝子）や遺伝子B（イネいもち病菌にはあり、アワいもち病菌には存在しない遺伝子）を探せば、宿主特異性の原因遺伝子を見つけられるかもしれません。

　今回の解析では、イネいもち病菌にはあり、アワいもち病菌にはない遺伝子を探し、遺伝子Bの候補をピックアップします。

## 遺伝子のカバー率の計算

　ここでは、遺伝子カバー率を百分率（％）で取得します。

```

```

#### 解答例

````
# イネいもち病菌のシーケンスリードがカバーしている塩基数(bp)
oryza_cover_bp = ine_cover_rate["Oryza_islolate_cover_bp"] 
# アワいもち病菌のシーケンスリードがカバーしている塩基数(bp)
setaria_cover_bp = ine_cover_rate["Setaria_islolate_cover_bp"]
# 遺伝子長(bp)
gene_length = ine_cover_rate["Gene_length"]

# イネいもち病菌の遺伝子カバー率
ine_cover_rate["Oryza_islolate_cover_rate"]  = ( oryza_cover_bp / gene_length )*100 

# アワいもち病菌の遺伝子カバー率
ine_cover_rate["Setaria_islolate_cover_rate"] = ( setaria_cover_bp / gene_length)*100 

# 確認
ine_cover_rate
````

　つぎに、2つのいもち病菌間でカバー率が大きく異なる遺伝子を探していきましょう。

## まとめ
　今回の実習では、**イネ菌とアワ菌のあいだで遺伝子がある・なしの比較**をおこないました。 

　おもに以下の流れで解析しました。
0.  イネいもち病菌とアワいもち病菌のシーケンスをゲノム配列にマッピング  <small>※ 説明のみ</small>
1. 各遺伝子において、シーケンスリードがカバーしている塩基数を取得 <small>※ 説明のみ</small>
2. 各遺伝子の長さ、シーケンスリードの遺伝子カバー率を計算
3. 2つのいもち病菌間でカバー率に差がある遺伝子（片方のいもち病菌にのみ存在する遺伝子）を検出

　今回のように **Python** が使えると、自分の手でゲノムデータ解析をおこなえます。
