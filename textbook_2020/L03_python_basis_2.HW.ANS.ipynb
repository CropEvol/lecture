## 課題1

　下記のコードセルに追記して、for構文で1行ずつ取り出して表示するプログラムを完成させてください。

````
f = open("allele_count_mini2.txt")  # ファイルを開く



f.close()  # ファイルを閉じる
````

### 解答例

　`for`構文を使うと、`open()`で読み込んだテキストファイルを1行ずつ処理できるようになります。

```python
# 書き方
for 変数 in ファイル変数:
  処理1
  処理2
   :
```

## 課題2

　リスト化された行データのうち、後ろの2列（4番目と5番目）の数字データを取り出して表示してください。なお、ここでの"4番目"や"5番目"といった番号は、0番から始まるインデックスの数え方による番号です。


````
f = open("allele_count_mini2.txt")

for line in f:
  line = line.split()  # リスト化

f.close()
````

### 解答例


　`split()`により、文字列を分割できます。文字列の空白がある箇所で分割されて、リストデータに変換されます。

<small>※ 区切り文字を指定することも可能です。詳しくは[こちら](https://docs.python.org/ja/3/library/stdtypes.html?highlight=split#str.split)を参照してください。</small>

　さらに、リストデータから個々の要素を取り出すには、**インデックス**を指定します。

```python
# 書き方
リスト[インデックス]
```

## 課題3

　リスト化された行データの4番目と5番目の数字データを使い、下記の計算をおこなって、その計算結果を表示してください。なお、ここでの"4番目"や"5番目"といった番号は、0番から始まるインデックスの数え方による番号です。

```python
# リファレンスと同じ塩基をもつリード数の割合
4番目の数値 / (4番目の数値 + 5番目の数値)
```

````
f = open("allele_count_mini2.txt")

for line in f:
  line = line.split()  # リスト化

f.close()
````

### 解答例


　文字列を`split()`で分割してリストデータにしても、個々の要素は文字列として保持されています。

　数値計算をおこなうには、数字を「文字列」ではなく「数値」に変換する必要があります。

```python
# 文字列 => 数値の変換
int("<数字>")
```

　数値変換後、各種の演算子記号（プラス記号など）を使用して、数値計算をおこないます。

```python
# 足し算: 半角のプラス記号を使う
数値 + 数値
# 引き算: 半角のマイナス記号を使う
数値 - 数値
# 掛け算: 半角のアスタリスク記号を使う
数値 * 数値
# 割り算: 半角のスラッシュ記号を使う
数値 / 数値
```

````
f = open("allele_count_mini2.txt")

for line in f:
  line = line.split()      # リスト化
  c1 = int(line[4])      # 4番目の数字データ; 数値変換
  c2 = int(line[5])      # 5番目の数字データ; 数値変換
  c1_freq = c1 / (c1 + c2)  # リファレンスと同じ塩基をもつリード数の割合
  print(c1_freq)              # 表示

f.close()
````
