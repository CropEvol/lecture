　前回「MutMap解析」 と前々回「Sliding Window解析」では、突然変異塩基の割合（**SNP-index**）が平均的に高いゲノム領域（原因遺伝子がある遺伝子座）の探索をおこないました。

　いくつ以上のSNP-indexであれば、"高い"とみなせるのでしょうか？　今回、SNP-indexシミュレーションをおこない、実験で得られたSNP-indexが高いか低いかをより客観的に判定できるようにしてみましょう。

## 突然変異塩基が得られる期待確率

　各SNP座のSNP-indexは、リファレンス塩基と突然変異塩基のそれぞれの個数を使って、突然変異塩基の割合を調べることで計算できました。

$$
SNP\text{-}index = \frac{ALT}{REF + ALT} \\
$$
上式のREFは「リファレンス塩基の個数」で、ALTは「突然変異塩基の個数」です。


　つぎに、突然変異塩基が得られる期待確率を考えてみましょう。親系統の対立遺伝子（アリル; ここでは、リファレンス塩基と突然変異塩基）は、F2集団の各個体に約50%の確率で遺伝されます。突然変異に無関係なSNP座であれば、突然変異形質をもつF2個体集団でも同じことが言えます。言い換えると、**突然変異形質に無関係なSNP座においては、突然変異形質のF2個体集団からランダムにいくつかの塩基を抽出すると、リファレンス塩基と突然変異塩基がおよそ1:1の割合で得られます。**

<img src="https://lh3.googleusercontent.com/pw/ACtC-3f-BCm9R6-PcClk86LW0zFjo2YRAr9gUOjAdNH8W0Cs_rky7EhITPtN2a9cb9yD0hvC1h8Kg8glp_5AGDItnDl6ZNw2-FPZ2vscYrXSthXkLORnPf2TYsRjjtwOHlQhBJKihHzzJU35cPjAJB9s9PQ1=w1747-h1037-no?authuser=0" alt="expected_allele_freq" height="280px">

　したがって、「突然変異塩基が得られる期待確率」は次のとおりです。

$$
p_{ALT} = 0.5
$$

「リファレンス塩基が得られる期待確率」も次のとおりです。

$$
p_{REF} = 1 - p_{ALT} = 0.5
$$

　今回おこなうシミュレーションでは、「突然変異塩基が得られる期待確率」を設定値（**パラメータ Parameter**）にして、**突然変異形質に無関係なSNP座で発生するSNP-indexのシミュレーション**をおこないます。

## SNP-indexのシミュレーションの手順

SNP-indexのシミュレーションの手順を確認しましょう。
1. 帰無仮説とパラメータを設定する
  - 帰無仮説: 実測SNP-indexは、突然変異形質に無関係なSNP座のSNP-index
  - パラメータ: 突然変異塩基発生確率 $p_{ALT}=0.5$  
<small>※ 実際のMutMapでは、 $p_{ALT}$　も確率的に発生させます。今回は、簡略化のために、$p_{ALT}=0.5$　の固定値を設定します。</small>
1. 1の設定のもとで任意個数の塩基を発生させて、擬似SNP-indexを得る。
1. 2の操作を任意の回数繰り返す。
1. 擬似SNP-indexの分布（帰無分布）を調べる。
1. 実測SNP-indexが帰無分布でどこに位置するのか調べて、帰無仮説にしたがうか判定する。
  - 頻繁に得られるSNP-indexのとき、帰無仮説に従う（帰無仮説を棄却しない）。→突然変異形質に無関係なSNP座
  - めったに得られないSNP-indexのとき、帰無仮説を棄却する。→突然変異形質に関係するSNP座  
<small>※ 多くの場合、帰無分布の端 5%未満（もしくは1%未満）に含まれる場合に「めったに得られないSNP-index」と判定します。</small>

```python
# 1個の塩基をシミュレーションで発生させる
import numpy as np
np.random.binomial(p=0.5, n=1)
```

<small>※ NumPyと呼ばれる数値計算ライブラリを使います。このライブラリは、行列計算、ランダムな数値の生成、数列の生成など、多様な機能をもっています。ゲノムデータ解析でもよく使うライブラリのひとつです。</small>



　`binomial()`の`p`には、突然変異塩基発生確率 $p_{ALT}=0.5$ を指定します。 また、`n`には、1個の塩基を発生させたいので`1`を指定します。

　次のコードセルを実行して、1個の塩基を発生させてみましょう。

　実行すると、突然変異塩基が何個得られたか（ここでは、0か1のどちらか）が返ってきます。

- **0 = 突然変異塩基が0個得られた（すなわち、リファレンス塩基が1個得られた）**
- **1 = 突然変異塩基が1個得られた**


````
# Numpyライブラリの準備
import numpy as np
# 1個の塩基を発生させるシミュレーション
simu_alt = np.random.binomial(p=0.5, n=1)
# シミュレーション結果の出力
print(simu_alt)
````

### 二項分布`binomial`

　ここで使用した`binomial`という関数は、二項分布から任意の個数の値をサンプリングする関数です。二項分布の詳細は省きますが、簡単に言えば、コイントスを任意回数繰り返して表が出た数を調べることができる関数です。

　`binomial`に与える`p`の値を変えることで、トスするコインを自由に設定できます。

- 公平なコイン: 表が50%の確率で出るコイン
- 歪んだコイン: 表が80%の確率で出るコイン
- 表しか出ないコイン: 表が100%の確率で出るコイン

### 2-2. 実測SNP-indexの位置と仮説検定

　「SNP-indexのシミュレーションの手順」に書いていた文章を再び掲載します。
> 5. 実測SNP-indexが帰無分布でどこに位置するのか調べて、帰無仮説にしたがうか判定する。
  - 頻繁に得られるSNP-indexのとき、帰無仮説に従う（帰無仮説を棄却しない）。→突然変異形質に無関係なSNP座
  - めったに得られないSNP-indexのとき、帰無仮説を棄却する。→突然変異形質に関係するSNP座  
<small>※ 多くの場合、帰無分布の端 5%未満（もしくは1%未満）に含まれる場合に「めったに得られないSNP-index」と判定します。</small>

　ここでは、まず、擬似SNP-indexの右端5%に相当するSNP-indexの値を得てみましょう。そして、実測SNP-indexがその右端5%の領域に含まれるか調べてみましょう。

　次のコードセルを実行すると、**1000個**の擬似SNP-indexとその分布を得ることができます。

````
# 擬似SNP-indexの準備
simu_alt = np.random.binomial(p=0.5, n=18, size=1000)
simu_index = simu_alt / 18

# 擬似SNP-indexの確認
# print(simu_index)

# 擬似SNP-indexの分布（密度分布プロット）
import seaborn as sns
sns.displot(simu_index, kde=True)
````

100個の擬似SNP-indexから「95パーセンタイル（上側5%点）」を得ると、　その値が右端5%の**閾値（Threshold）**に相当します。

## 3. 全SNP座のシミュレーション

　最後に、全SNP座についてシミュレーションをおこない、実測値とシミュレーション値のSliding window解析結果をグラフに描画しましょう。

<small>※ コードの解説はおこないません。</small>

````
# --- 実測データ ---
pos      = df["POS"]         # 実測POS
snpindex = df["SNP_index"]   # 実測SNP_index

# --- シミュレーション ---
# SNP座のシミュレーション関数
def one_locus_simulation(allele_num, trials=1000, q=0.95):
  simu_snp_index = np.random.binomial(n=allele_num, p=0.5, size=trials) / allele_num  # 擬似SNPindex
  q95_threshold = np.quantile(simu_snp_index, q=q)  # 閾値（戻り値）
  return q95_threshold
# 各座のシミュレーション
df["Threshold"] = (df["N_REF"]+df["N_ALT"]).apply(one_locus_simulation)

# --- Sliding window解析 ---
chrom_size = 23207287  # イネ第10番染色体のサイズ
win_size  = 1   * 1000 * 1000  # 区間サイズ: 1Mb
step_size = 0.2 * 1000 * 1000  # ステップサイズ: 0.2Mb

start = 1                    # 始点
end  = start + win_size - 1 # 終点
pos_win = []  # POSの区間平均用リスト
idx_win = []  # SNP_indexの区間平均用リスト
th_win  = []  # Thresholdの区間平均用リスト
while True:
  # 区間データの取得
  sub = df[ (df["POS"]>=start) & (df["POS"]<=end) ]
  # 区間平均の取得
  pos_win.append(sub["POS"].mean())
  idx_win.append(sub["SNP_index"].mean())
  th_win.append(sub["Threshold"].mean())
  # 区間の移動
  start = start + step_size
  end  = end  + step_size
  # ループ終了処理
  if end > chrom_size:
    break

# --- グラフ描画 ---
import matplotlib.pyplot as plt
plt.figure(figsize=[12,4])    # グラフサイズ
plt.scatter(pos, snpindex, alpha=0.5)   # 実測値SNP-index
plt.plot(pos_win, idx_win, color="green", label="Observed")  # 実測値のsliding window
plt.plot(pos_win, th_win, linestyle="--", color="red", label="Threshold")    # シミュレーション値のsliding window
plt.xlabel('Position (x 10 Mb)', fontsize=12)  # x軸ラベル
plt.ylabel('SNP-index', fontsize=12)           # y軸ラベル
plt.legend()
plt.show()
````

　全SNP座についてシミュレーションをおこなうと、第10番染色体右端の平均SNP-indexが閾値を超えるゲノム領域であることがわかりました。その領域は、「突然変異形質に無関係なゲノム領域」から得られる値にしては"有意"に高い値です。すなわち、「突然変異形質に関係するゲノム領域」とみなせそうです。

## まとめ

　今回、事前に設定した仮定（パラメータ）のもとでSNP-indexのシミュレーションをおこない、実験で得られた実測データを解釈する方法を学びました。

1. 帰無仮説とパラメータを設定する
1. 1の設定のもとで任意個数の塩基を発生させて、擬似SNP-indexを得る。
1. 2の操作を任意の回数繰り返す。
1. 擬似SNP-indexの分布（帰無分布）を調べる。
1. 実測SNP-indexが帰無分布でどこに位置するのか調べて、帰無仮説にしたがうか判定する。

　確率統計学的な解析を加えると、実測データの判断をより客観的におこなるようになります。

　シミュレーションは、様々なデータ解析（ベイズ推定など）で使われています。ゲノム解析を含むデータ解析でもよく利用されています。人間の手で1,000回、10,000回、100,000回の試行をおこなうのは大変ですが、コンピュータはそのような単純作業を高速かつ大規模におこなうことができます。この強力なツールを使いこなせると、研究・仕事の幅が広がるでしょう。
