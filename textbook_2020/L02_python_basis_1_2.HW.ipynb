## 課題2

　変数maizeのアミノ酸配列を、for構文とインデックスを使って、前から1文字ずつ順番に出力してください。

```
maize = "MAQPPQWKA"
```

#### ヒント

- アミノ酸配列の長さを`len()`で調べる
- インデックス(0からの数字)を順番に発生させる: `for i in range(アミノ酸配列の長さ):`
- `アミノ酸配列[インデックス]`で一文字取り出す
