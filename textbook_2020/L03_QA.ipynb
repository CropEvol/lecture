# サンプルプログラム1

## Q. 

print(line, line[4], line[5])と print(line,line[4:6])の意味が同じのように思いますが、なぜ出力は違うのでしょうか。

## A. 

- インデックスを指定する方法（`print(line[4], line[5])`）は、リストから個々の要素にアクセスしています。

- スライス（`print(line[4:6])`）は、リストの一部を切り出す方法です。切り出されたデータは、切り出し元と同じデータ型（リスト）が維持されます。

```
# 例
fruit = ["grape", "orange", "apple", "peach"]

# インデックスを使う
print(fruit[1], fruit[2])

# スライスを使う
print(fruit[1:3])
```

```
# 例）　インデックスを使って、リストから個々の要素を取り出す
f = open("allele_count_mini2.txt")

for line in f:
  line = line.split()  
  print(line[4], line[5]) # リストのインデックスを指定

f.close()
```

```
# 例）　スライスを使って、リストから範囲指定で要素を取り出す
f = open("allele_count_mini2.txt")

for line in f:
  line = line.split()  # リスト化
  print(line[4:6])

f.close()
```

# サンプルプログラム2

## Q.

複数の要素を取り出すときに、print(line, line[4], line[5])とするべきところを、print(line, line[4, 5])としてしまいエラーとなったのですが、複数の要素を取り出す方法は、この「print(line, line[4], line[5])」という１通りなのか疑問に思いました。

## A.

リストのスライスを使う方法があります。ただし、上述の「サンプルプログラム1」でも解説していますが、スライスで取り出されるデータはリストの状態です。



```
# 例）　join()を使う方法
f = open("allele_count_mini2.txt")

for line in f:
  line = line.split()  # リスト化
  print(line, line[4:6])    # スライスを使って複数要素を取り出す

f.close()
```
