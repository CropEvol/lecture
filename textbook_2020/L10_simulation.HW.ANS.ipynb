## 課題1

　次のコードセルに追記して、突然変異塩基発生確率$p_{ALT}=0.5$のもとで**20個**の塩基を発生させて、擬似SNP-indexを得てください。



````
import numpy as np

# 20個の塩基を発生させるシミュレーション
simu_alt = np.random.binomial()

# 擬似SNP-indexを得る
simu_index = 

print(simu_index)
````

#### 解答例



```
# ある事象が起こる確率pの試行をn回おこなったときに
# その事象が何回おこるかシミュレーションする
np.random.binomial(p=<事象発生確率>, n=<試行回数>)
```

`binomial()`に与える情報は、今回の場合、次のとおりです。
- 事象発生確率 = 突然変異塩基発生確率0.5
- 試行回数 = 20個の塩基を発生させる

`binomial()`を実行した結果、返ってくる情報は「20個の塩基のうち、突然変異塩基が得られた個数」です。

そこで、擬似SNP-indexを求めるために、`binomial()`から返ってきた値を20で割って、「20個中の突然変異塩基の割合」に変換します。


````
import numpy as np

# 20個の塩基を発生させるシミュレーション
simu_alt = np.random.binomial(p=0.5, n=20)

# 擬似SNP-indexを得る
simu_index = simu_alt /20

print(simu_index)
````

## 課題2

　次のコードセルに追記して、「突然変異塩基発生確率$p_{ALT}=0.5$のもとで**20個**の塩基を発生させる」という試行を**1000回**おこなって、1000個の擬似SNP-indexを得てください。

````
import numpy as np

# 「20個の塩基を発生させる」を1000回おこなうシミュレーション
simu2_alt = np.random.binomial()

# １０００個の擬似SNP-indexを得る
simu2_index = 

print(simu2_index)
````

#### 解答例

課題1の処理を1000回おこないたい場合、`binomial()`に`size`オプションを追加すれば簡単に実現できます。

```
# ある事象が起こる確率pの試行をn回おこなったときに
# その事象が何回おこるかシミュレーションする。
# そのシミュレーションをsizeで指定した反復回数おこなう
np.random.binomial(p=<事象発生確率>, n=<試行回数>, size=<反復回数>)
```

返ってくる情報は、反復回数分のシミュレーション結果のリスト状データです。今回の場合、「20個の塩基のうち、突然変異塩基が得られた個数」が1000個得られます。

あとは、その1000個の値それぞれを20で割り、1000個の擬似SNP-indexに変換します。

```
<binomial()から返ってくるデータ> / 20
```


````
import numpy as np

# 「20個の塩基を発生させる」を1000回おこなうシミュレーション
simu2_alt = np.random.binomial(p=0.5, n=20, size=1000)

# １０００個の擬似SNP-indexを得る
simu2_index = simu2_alt /20

print(simu2_index)
````

## 課題3

　課題2で発生させた1000個の擬似SNP-indexから、**95パーセンタイル**の値を取得してください。

　また、その次のコードセルを実行して、擬似SNP-indexの分布と95パーセンタイルの値を描画してください。

````
# 95パーセンタイル
th = np.quantile()
print(th)
````

````
# 擬似SNP-indexの分布（密度分布プロット）
import seaborn as sns
sns.displot(simu2_index, kde=True, label="pseudo-SNP-index distribution")

# 95パーセンタイルの直線
import matplotlib.pyplot as plt
try:
  plt.axvline(th, color="red", linestyle="--", label="95 percentile")
except:
  0

plt.legend()
plt.show()
````

#### 解答例

1000個の擬似SNP-indexから95パーセンタイル（小さい順に並べて95%番目の値）を求める方法は、

```
np.quantile(<1000個の擬似SNP-indexデータ>, q=0.95)

# `q`に指定する値を変更すれば、指定のパーセンタイルの値を取得できます。
# Median(中央値、50パーセンタイル)や99パーセンタイルなど
```



````
# 95パーセンタイル
th = np.quantile(simu2_index, q=0.95)
print(th)
````

````
# 擬似SNP-indexの分布（密度分布プロット）
import seaborn as sns
sns.displot(simu2_index, kde=True, label="pseudo-SNP-index distribution")

# 95パーセンタイルの直線
import matplotlib.pyplot as plt
try:
  plt.axvline(th, color="red", linestyle="--", label="95 percentile")
except:
  0

plt.legend()
plt.show()
````
