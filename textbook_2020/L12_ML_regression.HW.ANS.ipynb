決定係数 $R^2$ で学習済みのモデルを評価する


- $R^2$が0に近いほどデータの当てはまりが悪いモデル
- $R^2$が1に近いほどデータの当てはまりが良いモデル
- トレーニングデータとテストデータの両方で高い値を示すモデルがよい
