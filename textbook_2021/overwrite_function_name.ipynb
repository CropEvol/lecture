# 関数名の上書きエラーについて

　まずは、下のprint関数やlen関数が正常に動くことを確認してください。

```python
print("ATGC")
```

```python
print(len("ATGC"))
```

　次のコードセルを実行すると、関数と同じ名前の変数が作成されます。

```python
print = "AAAAAAAAA"
len   = "CCCCCCCCCC"
```

　上のコードセルを実行したあと、以下のコードを実行すると、エラーが発生します。

```python
print("ATGC")
```

```python
print(len("ATGC"))
```

　これは、関数と同じ名前の変数を作成すると、元々定義されていた関数名が変数名に上書きされ、関数として使用できなくなるためです。

　Pythonに標準で定義されている関数（例えば、以下の名前の関数）は多数あります。できるかぎり関数と同じ名前の変数は作らないようにするのがよいでしょう。
- print
- len
- count
- list
- dict
- str
- int
- sum

　変数名を複合語にするなどすれば、関数名を上書きしてしまうことは少なくなるでしょう。

例:
- dna_len
- count_A
