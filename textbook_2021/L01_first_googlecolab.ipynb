# Google Colaboratoryの使い方

## はじめに
　「ゲノム情報解析入門」では、**Python** とよばれるプログラミング言語を使って、データ解析の実習をおこなっていきます。

　今回の実習では、Pythonプログラミングをおこなう前に、この **Google Colaboratory** というオンラインサービスの使い方を学びます。


### プログラミング ・ Python ・ Google Colaboratory


　**プログラミング** や **Python**、**Google Coraboratory** といったこれらは何なのか？　

- **プログラミング (Programming)**  
　規程された一連の動作をおこなうもののことを **プログラム (Program)** といいます。 **プログラミング** はプログラムを作る（書く）ことを指します。  
　よく似た言葉に **コード(Code)** や **コーディング（Coding）** があります。この講義では、プログラムとコードを、また、プログラミングとコーディングを同義語として扱っています。

- **Python**  
　プログラミング言語のひとつです。Pythonは、データ解析用の追加機能（パッケージ）が充実しており、データ解析の分野ではよく使われています。Python以外にも多くのプログラミング言語があります（C、C#、C++、Java、Ruby、PHP、Javascript、Rなど）。

- **Google Coraboratory**  
　Pythonプログラミングをおこなうための便利なオンラインサービスです。名称からわかるとおり、Google社が提供しているサービスです。  
　Google Coraboratoryという名称はやや長いので、以降では「Google Colab」や「Colab」といった省略名を使用します。



### 実習に関してのアドバイス


　多くの人にとって、PythonやGoogle Colabolatoryは、研究データ等を解析するための「道具」です。
 
　実習では、おもに道具の「使い方」（プログラムの基本構文や使用方法、など）を解説します。それを知っていなければ、自身の研究データなどに応用できることに気付けないからです。
 
　しかし、「使い方」に関しては、書籍やWebで調べれば、後からでも情報を得られるので、完璧に覚える必要はありません。それよりも、それを使えば「何ができるのか」を理解する方が今後の研究や仕事のためになるでしょう。 
 
　実習の取組み方としては、内容をパーフェクトに覚えるのではなく、Pythonプログラミングやデータ解析の「体験」を目標にするのが良いでしょう。


### ノートブックの「保存」と「ダウンロード」の違い

- 保存は、Googleドライブにノートブックファイルを残す操作です。
- ダウンロードは、手元のパソコンにノートブックファイルを残す操作です。

### Colabサーバー上のデータは保存されない

　Colabサーバーは、一定時間操作がないと、サーバーがOFFになります。それと同時に、サーバー上のファイルも削除されます。そのため、**Colabサーバー上にファイルを保存しておくことはできません**。

　編集内容を保存したい場合は、ノートブックファイルをGoogleドライブに「保存」するか、手元のパソコンに「ダウンロード」してください。
