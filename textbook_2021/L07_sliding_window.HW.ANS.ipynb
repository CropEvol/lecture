### Sliding window解析の概要

　Sliding window解析は、任意の値（ここではSNP-index）の区間平均をゲノム全域にわたって調べて、区間平均（推定される理論値）が極端に低くなる場所や高くなる場所を探す解析手法です。

<img src="https://lh3.googleusercontent.com/pw/ACtC-3euwLsOQQgwmqJWLWTeSROFjby5RYnNH_UdqTE9SEpefx_XkZILIKAdrOsq0bzT_7L6RZQFwAFgexYRUupN8c6EdZVUyNN96gSXsj6mTM8v26qaCZs2HCllVzuWexZZ2t8J_tAN80NbuloO79WWK5IR=w720-h405-no?authuser=0" alt="sliding_window_overview" height="300px">

　Sliding window解析は次の手順でおこないます。
1. 区間サイズとステップサイズ（次の区間への移動サイズ）、各区間の代表値を決める。  
<small>※ 代表値として「平均値」が使われることが多い。</small>
1. 1区間内に含まれるデータを取り出す。
1. 代表値を求める。
1. 区間をステップサイズ分だけ移動させる。
1. 手順2-4を繰り返し、全区間を調べる。

#### 解答例(1): 1区間目のコードを真似する

　各区間の範囲に注意して、1区間目のコードをまねて、2区間目〜6区間目のコードを記述すればOKです。

- 最初の区間(1区間目): 1 - 2,000,000 bp
- 2区間目: 400,001 - 2,400,000 bp
- 3区間目: 800,001 - 2,800,000 bp
- 4区間目: 1,200,001 - 3,200,000 bp
- 5区間目: 1,600,001 - 3,600,000 bp
- 6区間目: 2,000,001 - 4,000,000 bp

````
# Sliding window用リスト
x_win = []  # POSの平均値を入れるリスト
y_win = []  # SNP_indexの平均値を入れるリスト

# 各区間（1区間目）の平均値を調べる
sub = df[ (df["POS"]>=1) & (df["POS"]<=2000000) ]
p = sub["POS"].mean()
s = sub["SNP_index"].mean()

x_win.append(p) # リストへの追加
y_win.append(s)

# 2区間目〜6区画目までの各区間の平均値を調べる
# 2区間目
sub = df[ (df["POS"]>=400001) & (df["POS"]<=2400000) ]
p = sub["POS"].mean()
s = sub["SNP_index"].mean()

x_win.append(p) # リストへの追加
y_win.append(s)

# 3区間目
sub = df[ (df["POS"]>=800001) & (df["POS"]<=2800000) ]
p = sub["POS"].mean()
s = sub["SNP_index"].mean()

x_win.append(p) # リストへの追加
y_win.append(s)

# 4区間目
sub = df[ (df["POS"]>=1200001) & (df["POS"]<=3200000) ]
p = sub["POS"].mean()
s = sub["SNP_index"].mean()

x_win.append(p) # リストへの追加
y_win.append(s)

# 5区間目
sub = df[ (df["POS"]>=1600001) & (df["POS"]<=3600000) ]
p = sub["POS"].mean()
s = sub["SNP_index"].mean()

x_win.append(p) # リストへの追加
y_win.append(s)

# 6区間目
sub = df[ (df["POS"]>=2000001) & (df["POS"]<=4000000) ]
p = sub["POS"].mean()
s = sub["SNP_index"].mean()

x_win.append(p) # リストへの追加
y_win.append(s)


# リストの中身を確認
print(x_win)
print(y_win)
````

#### 解答例(2): 繰り返し構文（`for`構文）を使う方法1

 6回同じ処理を繰り返すために `for i in range(6):` を記述します。

 各ループ中に以下の処理をおこないます。
 - 区間データのとりだし
 - 区間平均の算出
 - 区間平均をリストに追加
 - 区間の更新


````
# Sliding window用リスト
x_win = []  # POSの平均値を入れるリスト
y_win = []  # SNP_indexの平均値を入れるリスト

# 区間範囲の初期値
start = 1
end  = 2000000

for i in range(6):
  # 区間データの取り出し
  sub = df[ (df["POS"]>=start) & (df["POS"]<=end) ]
  # 区間平均の算出・リストへの追加
  p = sub["POS"].mean()
  s = sub["SNP_index"].mean()
  x_win.append(p)
  y_win.append(s)
  # 区間の更新
  start = start + 400000
  end  = end  + 400000

# リストの中身を確認
print(x_win)
print(y_win)
````

#### 解答例(3): 繰り返し構文（`for`構文）を使う方法2


　少し高度な方法ですが、for構文を使う場合、変数`i`に入る数値データを使って、区間データの取り出しと区間の更新を同時におこなうことも可能です。

````
# Sliding window用リスト
x_win = []  # POSの平均値を入れるリスト
y_win = []  # SNP_indexの平均値を入れるリスト

for i in range(6):
  # 区間データの取り出し（変数iを使う）
  sub = df[ (df["POS"] >= i * 400000) & (df["POS"] <=  i * 400000 + 2000000) ]
  # 区間平均の算出・リストへの追加
  p = sub["POS"].mean()
  s = sub["SNP_index"].mean()
  x_win.append(p)
  y_win.append(s)

# リストの中身を確認
print(x_win)
print(y_win)
````

#### 解答例(4): 繰り返し構文（`while`構文）を使う方法


　`for`構文以外にも、`while`構文を使って、同様の処理をおこなうことも可能です。

````
# Sliding window用リスト
x_win = []  # POSの平均値を入れるリスト
y_win = []  # SNP_indexの平均値を入れるリスト

# 区間範囲の初期値
start = 1
end  = 2000000

# ループ回数を記録するカウンター
i = 1

while True:
  # 区間データの取り出し
  sub = df[ (df["POS"]>=start) & (df["POS"]<=end) ]
  # 区間平均の算出・リストへの追加
  p = sub["POS"].mean()
  s = sub["SNP_index"].mean()
  x_win.append(p)
  y_win.append(s)
  # 区間の更新
  start = start + 400000
  end  = end  + 400000

  # カウンターの繰り上げ
  i = i + 1
  if i >= 7:
    break

# リストの中身を確認
print(x_win)
print(y_win)
````
