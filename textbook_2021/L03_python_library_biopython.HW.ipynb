## 課題2

　サンプルデータ`NGS_sample_mini_2.fastq`はFASTQ形式ファイルです。

　次のコードセルには、Biopythonを使って、ファイルの中身をFASTQ形式データとして読み込み、塩基配列データを順番に取り出し、そのデータを表示するプログラムが書かれています。

　各塩基配列の名前と塩基配列本体を表示するプログラムに書き換えてください。


````
### BiopythonライブラリのSeqIOを準備する
from Bio import SeqIO

### ファイルを開く
f = open("NGS_sample_mini_2.fastq")

### BiopythonでFASTQファイルを読み込む
recs = SeqIO.parse(f, "fastq")

### 繰り返し処理: FASTQファイルから塩基配列データを順番に取り出し、表示する
for rec in recs:
  print(rec) # 表示

### ファイルを閉じる
f.close()
````
