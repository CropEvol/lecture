## 課題1: プログラムを実行する

　次のコードセルのプログラムを実行してください。

いずれかの方法で、下記のコードセルを実行する

- (Windowsの場合) Ctrl + Enter
- (MacOSの場合) control + Enter　または command + Enter

````
# DNA配列中の"A"の個数を調べるプログラム
dna = "GGCTTATTAGCTGAA" 
cnt = dna.count("A")
print(cnt)
````

## 課題2: プログラムを一部書き換えて、実行する

　課題1のプログラムの`cnt = dna.count("A")`を、次のように変更して実行してください。
 
```python
cnt = dna.count("GCT")
```

下記のコードセルの `"A"` を `"GCT"` に書き換える

````
# 下記のプログラムを書き換えて実行する
dna = "GGCTTATTAGCTGAA" 
cnt = dna.count("A")
print(cnt)
````

````
# 書き換え後のプログラム
dna = "GGCTTATTAGCTGAA" 
cnt = dna.count("GCT")
print(cnt)
````

## 課題3: 共有リンクを発行する
