## サンプルデータの説明

　[サンプルデータ](https://raw.githubusercontent.com/CropEvol/lecture/master/data/pandas_hw_dataset_mini.txt) は、101行6列のテーブル状（表状）のデータで、**カンマ区切りファイル (Comma-Separated Values file; CSV file)** として記述されています。

　このCSVファイルには次のような情報が書かれています。
- 最初の行は、各列の名前が書かれた**ヘッダー行 (Header line)**。
- 2行目以降の各行は、各SNP座の情報
- 各列は、
  - 染色体名
  - 染色体中の塩基座位の場所
  - リファレンスの塩基
  - リファレンスとは異なる塩基
  - リファレンスと同じ塩基をもつリード数
  - リファレンスと異なる塩基をもつリード数

## サンプルデータのダウンロード


　課題をおこなう前に、下記のコードセルを実行してください。[サンプルデータ](https://raw.githubusercontent.com/CropEvol/lecture/master/data/pandas_hw_dataset_mini.txt)のダウンロードと読み込みがおこなわれます。

　読み込まれたデータは、データフレーム変数`df`に格納されます。

````
"""
＊重要＊
最初にこのセルを実行してください。
この課題で使うサンプルファイルをダウンロードします。
"""
# サンプルデータをダウンロードする
!wget -O hw_dataset_mini.txt https://raw.githubusercontent.com/CropEvol/lecture/master/data/pandas_hw_dataset_mini.txt

# pandasでサンプルデータを読み込む
import pandas as pd
df = pd.read_csv("hw_dataset_mini.txt", sep=",", header=0)

# 各SNPアリル頻度の計算
df["F_REF"] = df["N_REF"]/(df["N_REF"] + df["N_ALT"])
df["F_ALT"] = 1 - df["F_REF"]
# 表示
df
````

## 課題1

　2つの列データ`F_REF`と`F_ALT`を使って、下記の計算をおこなってください。また、その計算結果を新しい列`Hetero`としてデータフレームに追加してください。

$$ 2 \times F\_REF \times F\_ALT $$

<small>※ この式は、各遺伝子座の「ヘテロ接合度の期待値」を求める式です。</small>


#### 解答例

````
# 2つの列データを計算し、新しい列として追加
df["Hetero"] = 2 * df["F_REF"] * df["F_ALT"]

# 確認
df
````

新規データフレーム変数 = データフレーム[条件式]

## 課題5

<small>※ この課題は回答不要です。</small>

　**MutMap法** について調べてください。

#### 解答例

　**MutMap法**は、突然変異形質を引き起こす原因変異の位置(ゲノム上の座標)を特定する方法です。

　MutMap法の手順は次のとおりです。
1. ある系統に対して、化学的処理（EMS処理）または物理的処理（ガンマ線照射処理）をおこない、ゲノム中の数カ所に人為的に**突然変異を誘発**させた系統を作ります。  
2. その突然変異系統と元の系統（オリジナル系統）のあいだで交配し、F2集団を作ります。F2集団は、遺伝的組換えにより、**ゲノムがシャッフルされた集団**です。
3. 交配F2集団のうち、突然変異形質をもつ個体からDNAを抽出し、ひとまとめにしてシーケンシング（**バルクシーケンシング**）をおこないます。また、オリジナル系統のシーケンシングもおこないます。
4. オリジナル系統のゲノム配列に対してバルクシーケンシングリードをアライメントし、ゲノム配列とバルクシーケンシングリードのあいだで異なる塩基（突然変異で誘発された塩基; **突然変異塩基**）を見つけます。
5. 見つけた各塩基座位について、バルクシーケンシングリード中の**突然変異塩基の割合（SNP-index）**を調べます。
6. 親系統からF2世代への塩基（アリル; ここではオリジナル系統の塩基と突然変異塩基）の伝達は、基本的にランダムに（約半々の割合で）伝わります。そのため、多くの塩基座位では、SNP-indexは約0.5付近の値を示します。  
　しかし、変異形質をもつF2個体は、その変異形質を引き起こした原因変異を共通に持っているはずなので、**原因変異箇所ではSNP-indexは1.0**を示します。  
　したがって、MutMap法では、ゲノム中からSNP-indexが1.0を示す場所を探し、そのあb書を原因変異として特定します（シーケンシングエラーが起こりうることも考慮すると、SNP-indexが1.0に近い場所を探します。）。
