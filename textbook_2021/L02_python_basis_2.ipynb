# Pythonの基礎・補足　データ型について


　Pythonには、様々な種類のデータがあります。その種類のことを **データ型** と呼びます。

　ここでは、基本的なデータ型を確認しましょう。
- 文字列
- 数値（整数、小数点数）
- ブーリアン値
- リスト
- 辞書（ディクショナリ）

## 文字列

　**文字列（string）**は、Pythonの基礎でも出てきた型です。英数字や記号の文字の並びをクオーテーション記号（`'`や`"`）で挟むと、そのデータ型になります。

　なお、データ型を確認したい場合は、type関数を使います。
```python
type(調べたいデータ)
```



````
# 例
s = "ATAGAATTCT"

# データ型を確認
print(s, type(s))
````

## 数値（整数、小数点数）

　クオーテーション記号で挟まない半角数値は、**数値**と呼ばれるデータ型になります。

　数値はおもに次の二つのデータ型に分かれます。
- **整数（int）**: `0`、`1`、`-2`、`10`、`10000`など
- **小数点数（float）**: `0.1`、`-1.2`、`2.3`、`10.0`、`100.0`など

　Pythonの基礎では、`len`関数から返ってくるデータが「整数」でした。



````
# 例
n1 = 10000000000
n2 = 1.0
L = len("ATAGAATTCT") # 文字列の長さ

# データ型を確認
print(n1, type(n1))
print(n2, type(n2))
print(L, type(L))
````

## ブーリアン

　**ブーリアン（boolean）**は、あまり聞きなれない言葉かもしれませんが、理解が難しいデータ型ではありません。簡単に言えば、「オンとオフ」、「1と0」といった二値状態をもつデータ型です。

　Pythonでは、ブーリアンの値として、`True`と`False`が使われています。

　Pythonの基礎では、条件式から返ってくるデータが「boolean」でした。



````
# 例
b1 = True
b2 = False
cond = len("ATAGAATTCT")==10 # 条件式: 文字列の長さが10と同じかどうか

# データ型を確認
print(b1, type(b1))
print(b2, type(b2))
print(cond, type(cond))
````

## リスト

　**リスト（list）**は、0個以上のデータが集まってひとつのデータになった型です。今後の実習でもたびたび登場するはずです。

```python
# リストの作り方
リスト変数 = [値0, 値1, 値2, 値3, 値4]
```

````
# 例
fruits = ["apple", "lemon", "orange", "grape"]
nums = [1, 2, 3, 4, 5]
various = ["a", 1, True, False, fruits]
blank_list = []

# データ型を確認
print(fruits, type(fruits))
print(nums, type(nums))
print(various, type(various))
print(blank_list, type(blank_list))
````

　リストの各要素にアクセスするには、リストの各要素の位置番号（**インデックス、添字**）を指定します。

　文字列の位置番号と同じく、最初の要素から0番、1番、2番、・・・と位置番号が振られています。

````
# 例
fruits = ["apple", "lemon", "orange", "grape"]

# "orange"を取り出す
print(fruits[2])
print(type(fruits[2]))
````

## 辞書（ディクショナリ）
　**辞書（ディクショナリ; dictionary）**も、リストと同様に複数のデータが集まってひとつのデータになった型です。

　リストと異なる点は、各要素は**キー（key）**と**値（value）**がセットになっていることです。

```python
# 辞書（ディクショナリ）の作り方
辞書変数 = { キー1 : 値1, キー2 : 値2 }
```

````
# 例
signals = { "off":False, "on":True, 0:False, 1:True }
univ = { "ku":"Kyoto University", "ou":"Osaka University" }
blank_dict = {}

# データ型を確認
print(signals, type(signals))
print(univ, type(univ))
print(blank_dict, type(blank_dict))
````

　辞書（ディクショナリ）の各要素にアクセスするには、「キー」を指定します。

````
# 例
univ = { "ku":"Kyoto University", "ou":"Osaka University" }

# "Kyoto University"を取り出す
print(univ['ku'])
````
