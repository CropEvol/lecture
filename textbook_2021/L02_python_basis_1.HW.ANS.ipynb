## 課題5

<small>※ これは、取り組み必須課題ではありません。</small>


　変数`maize`のアミノ酸配列には文字「A」（アラニン）がいくつ含まれているか調べてください。

　講義で習っていない関数を使っても構いません。

````
maize = "MAQPPQWKAMYQYVTRRAHDGCARVEESVAAARGALATPMVLDT"
````

### 解答例1: カウンタ用の変数を使う方法


````
maize = "MAQPPQWKAMYQYVTRRAHDGCARVEESVAAARGALATPMVLDT"

# カウンタ用の変数（初期値:0）
count_A = 0

# １文字ずつ取り出す
for i in maize:
  # 取り出した文字が「A」のとき、カウンタ用変数を+１
  if i == "A":
    count_A = count_A + 1

# カウンタ用変数の値を表示
print(count_A)
````

### 解答例2: `count`関数を使う方法

```python
# 書き方
対象の文字列.count(検索文字列)
# 例
"kyoto".count("o")
```

````
maize = "MAQPPQWKAMYQYVTRRAHDGCARVEESVAAARGALATPMVLDT"

# count関数で文字「A」の個数を調べる
count_A = maize.count("A")

# カウンタ用変数の値を表示
print(count_A)
````
