## サンプルデータの説明

　[サンプルデータ](https://raw.githubusercontent.com/CropEvol/lecture/master/data/pandas_hw_dataset_mini.txt) は、101行6列のテーブル状（表状）のデータで、**カンマ区切りファイル (Comma-Separated Values file; CSV file)** として記述されています。

　このCSVファイルには次のような情報が書かれています。
- 最初の行は、各列の名前が書かれた**ヘッダー行 (Header line)**。
- 2行目以降の各行は、各SNP座の情報
- 各列は、
  - 染色体名
  - 染色体中の塩基座位の場所
  - リファレンスの塩基
  - リファレンスとは異なる塩基
  - リファレンスと同じ塩基をもつリード数
  - リファレンスと異なる塩基をもつリード数
