## 課題1

　サンプルデータ[`mini_genome.fasta`](https://raw.githubusercontent.com/CropEvol/lecture/master/data/mini_genome.fasta)は、仮想生物のゲノム配列です（9本の染色体chromosome_01〜chromosome09の配列）。

　この仮想生物のゲノムサイズ（各染色体の長さの合計値）を調べてください。

````
### BiopythonのSeqIOを準備
from Bio import SeqIO

### ファイルを開く
f = open("mini_genome.fasta")

### FASTA形式で読み込み
recs = SeqIO.parse(f, "fasta")

### 1データ（1染色体）ずつ取り出す
for rec in recs:
  print(rec.id)
  print(rec.seq)

### ファイルを閉じる
f.close()
````

### 解答例: `sum`関数を使わない方法

重要なポイント: 
- ゲノムサイズ用の変数を事前に準備する
  - 11行目: `chrom_size = 0`
- 染色体の長さを取得した際に、上記の変数にその値を追加する（変数の値を更新する）
  - 19行目: `chrom_size = chrom_size + len(rec.seq)`

````
### BiopythonのSeqIOを準備
from Bio import SeqIO

### ファイルを開く
f = open("mini_genome.fasta")

### FASTA形式で読み込み
recs = SeqIO.parse(f, "fasta")

### ゲノムサイズ用の変数（初期値0）
chrom_size = 0

### 1データ（1染色体）ずつ取り出す
for rec in recs:
  #print(rec.id)
  #print(rec.seq)

  ### 調べた染色体の長さをゲノムサイズ用変数に追加
  chrom_size = chrom_size + len(rec.seq)

### 結果を表示
print(chrom_size)

### ファイルを閉じる
f.close()
````

### 解答例: `sum`関数を使う方法

重要なポイント: 
- 染色体の長さを一時保管するためのリストを事前に準備する
  - 11行目: `chrom_size_list = []`
- 染色体の長さを取得した際に、一時保管用リストにその値を追加する
  - 19行目: `chrom_size_list.append(len(rec.seq))`
- `sum`関数でリストの合計値を得る
  - 22行目: `print(sum(chrom_size_list))`


```python
# リストに値を追加する方法1
# （append関数でリストへの値の追加をおこなえる）
リスト.append(値)

# 例1
fruits = ["apple", "orange"]
fruits.append("grape")
print(fruits) # 出力結果=> ["apple", "orange", "grape"]
```

```python
# リストに値を追加する方法2
# （リストの足し算で、リストの結合をおこなえる）
リスト.append(値)

# 例2
fruits = ["apple", "orange"]
fruits = fruits + ["grape"]
print(fruits) # 出力結果=> ["apple", "orange", "grape"]
```

````
### BiopythonのSeqIOを準備
from Bio import SeqIO

### ファイルを開く
f = open("mini_genome.fasta")

### FASTA形式で読み込み
recs = SeqIO.parse(f, "fasta")

### 各染色体の長さを一時保管するためのリスト変数（初期値0）
chrom_size_list = []

### 1データ（1染色体）ずつ取り出す
for rec in recs:
  #print(rec.id)
  #print(rec.seq)

  ### 調べた染色体の長さをリストに追加
  chrom_size_list.append(len(rec.seq))

  ### 調べた染色体の長さをリストに追加（以下でもOK）
  #chrom_size_list = chrom_size_list + [len(rec.seq)]

### リスト内の値の合計値を調べる
print(sum(chrom_size_list))

### ファイルを閉じる
f.close()
````

## 課題3

　仮想生物の各染色体のGC含量（塩基GとCの割合）を調べてください。

```
# ひとつの染色体中のGC含量
GC含量 = (塩基Gの個数 + 塩基Cの個数) / 染色体の長さ
```



````
### BiopythonのSeqIOを準備
from Bio import SeqIO

### ファイルを開く
f = open("mini_genome.fasta")

### FASTA形式で読み込み
recs = SeqIO.parse(f, "fasta")

### 1データ（1染色体）ずつ取り出す
for rec in recs:
  print(rec.id)
  print(rec.seq)

### ファイルを閉じる
f.close()
````

### 解答例

````
### BiopythonのSeqIOを準備
from Bio import SeqIO

### ファイルを開く
f = open("mini_genome.fasta")

### FASTA形式で読み込み
recs = SeqIO.parse(f, "fasta")

### 1データ（1染色体）ずつ取り出す
for rec in recs:
  #print(rec.id)
  #print(rec.seq)
  
  ### Gの個数
  count_G = rec.seq.count("G")
  ### Cの個数
  count_C = rec.seq.count("C")
  ### 染色体の長さ
  chrom_size = len(rec.seq)
  ### GC含量
  GC_content = (count_G + count_C) / chrom_size

  ### 結果を表示
  print(rec.id, GC_content)

### ファイルを閉じる
f.close()
````
