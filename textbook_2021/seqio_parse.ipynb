# PHRED形式データ（.phd）の読み込み

```python
### Biopythonのインストール
!pip install biopython

### サンガーシーケンスのPHRED形式ファイル
!wget -O 'sanger_seq_sample.phd'   https://raw.githubusercontent.com/CropEvol/lecture/master/data/sanger_seq_sample.phd
```

```python
### BIopythonライブラリ
from Bio import SeqIO

### Genbankファイルを開く
f = open("sanger_seq_sample.phd")

### Biopythonで読み込む; 1データのみ取り出す
rec = next(SeqIO.parse(f, "phd"))
#print(rec)  # 確認

### letter_annotationsのキーと内容を確認
print(rec.letter_annotations.keys())
print('phred_quality:', rec.letter_annotations['phred_quality'])
print('peak_location:', rec.letter_annotations['peak_location'])

### ファイルを閉じる
f.close()
```

# Genbank形式データ（.gb）の読み込み

サンプルデータのソース: https://www.ncbi.nlm.nih.gov/nuccore/NC_012920

```python
### Biopythonのインストール
!pip install biopython

### Genbank形式ファイル (ヒトのミトコンドリアゲノム)
!wget -O 'NC_012920.gb'   https://raw.githubusercontent.com/CropEvol/lecture/master/data/NC_012920.gb
```

```python
### BIopythonライブラリ
from Bio import SeqIO

### Genbankファイルを開く
f = open("NC_012920.gb")

### Biopythonで読み込む; 1データのみ取り出す
rec = next(SeqIO.parse(f, "genbank"))
print(rec)  # 確認

### ファイルを閉じる
f.close()
```
