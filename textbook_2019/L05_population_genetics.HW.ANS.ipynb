　ここに3つの塩基配列があります。それぞれのペア（seq1-seq2、seq1-seq3、seq2-seq3）のSegregating siteの数を調べてください。
 
実習テキストの関連箇所: 
- Segregating site ($S$)

````
seq1 = "ACCTGGTTGATCCTGCCAGTAGTCATATGCTTGTCTCAAAGATTAAGCCATGCATG"
seq2 = "ACCTGGTTGATCCTGGCGGTAGACATATGCTAGTCTCAAAGAATAACCCATGCATG"
seq3 = "ACCTCGTTGATCCTGCCAGTAGTCATATGCTTGTCTCAAAAATTAACCCATGCAAG"

# 以下にコードを追記してください。

````

### 解答例

````
seq1 = "ACCTGGTTGATCCTGCCAGTAGTCATATGCTTGTCTCAAAGATTAAGCCATGCATG"
seq2 = "ACCTGGTTGATCCTGGCGGTAGACATATGCTAGTCTCAAAGAATAACCCATGCATG"
seq3 = "ACCTCGTTGATCCTGCCAGTAGTCATATGCTTGTCTCAAAAATTAACCCATGCAAG"

def count_seg(seq_a, seq_b):  #count_segという名前の関数を設定する
  S = 0  #segregating siteの初期値
  for nuc1, nuc2 in zip(seq_a, seq_b):  #seq_aの塩基配列であるnuc1とseq_bの塩基配列であるnuc2に関して、ループ関数を用いる
    if nuc1 != nuc2:  #nuc1とnuc2が異なる場合
      S = S + 1  #segregating siteのカウントを１つ増やす
  print(S)  #最終的に求まったsegregating siteの数を出力する

count_seg(seq1, seq2)  #上で作成した関数に実際の配列であるseq1, seq2を代入する
count_seg(seq1, seq3)
count_seg(seq2, seq3)
````

## Exercise 3
　次のセルのリストには、青森県でサンプリングしたオニドコロの一部ゲノム配列について、基準配列と同じ塩基の数（ref）と、異なる塩基の数（alt）を調べた結果がまとめられています。このデータから、Nucleotide diversity ($\pi$) を求めてください。
 
実習テキストの関連箇所: 
- Nucleotide diversity ($\pi$)

````
ref = [11,10,17,16,17,20,15,15,14,15,15,13,13,14,13,20,20,7,15,15,18,18,11,12,11,11,6,9,9,10,11,6,20,19,15,17,13,15,13,16,20,20,20,13,15,20,13,20,20,14,15,12,17,20,20,11,20,20,14,14,12,16,20,15,14,13,12,17,20,20]
alt = [9,10,3,4,3,0,5,5,6,5,5,7,7,6,7,0,0,13,5,5,2,2,9,8,9,9,14,11,11,10,9,14,0,1,5,3,7,5,7,4,0,0,0,7,5,0,7,0,0,6,5,8,3,0,0,9,0,0,6,6,8,4,0,5,6,7,8,3,0,0]


# 以下にコードを追記してください。

````

### 解答例

````
ref = [11,10,17,16,17,20,15,15,14,15,15,13,13,14,13,20,20,7,15,15,18,18,11,12,11,11,6,9,9,10,11,6,20,19,15,17,13,15,13,16,20,20,20,13,15,20,13,20,20,14,15,12,17,20,20,11,20,20,14,14,12,16,20,15,14,13,12,17,20,20]
alt = [9,10,3,4,3,0,5,5,6,5,5,7,7,6,7,0,0,13,5,5,2,2,9,8,9,9,14,11,11,10,9,14,0,1,5,3,7,5,7,4,0,0,0,7,5,0,7,0,0,6,5,8,3,0,0,9,0,0,6,6,8,4,0,5,6,7,8,3,0,0]

nume = 0  # "(基準と同じ塩基の数)　* (基準と違う塩基の数) / 配列の組み合わせ" の分子の初期値を設定する
for nuc_ref, nuc_alt in zip(ref, alt):  #refの塩基であるnuc_refとaltの塩基であるnuc_altについてループ関数を用いる
  nume = nume + nuc_ref * nuc_alt  #(基準と同じ) * (基準と違う) の総和を計算する
n = ref[0] + alt[0]  #染色体(配列)の総数を求める
pi = nume / (n * (n - 1) / 2) #numeと配列の総数からpiを計算する
print(pi)
````
