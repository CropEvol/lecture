## 課題2: プログラムを書いて、実行する

　すぐ下に新しい「コードセル」を作成し、次のプログラムを書いて、実行してください。
 
```python

word = "Bioinformatics" 
cnt = word.count('i')
print("i:", cnt)
```

### 解答例

1. 上記のプログラムをコードセルに記述する

````
# 文字列中の"i"の個数を調べるプログラム
word = "Bioinformatics" 
cnt = word.count('i')
print("i:", cnt)
````
