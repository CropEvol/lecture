## ファイルの読み込み方法

　テキストファイルの基本的な処理方法は次の通りです。
- ファイルを開く（open関数）
- 一行ずつ読み込む（for構文）
  - 末尾の改行文字を取り除く（rstrip関数）
  - 何らかの処理をする; 以下では塩基配列の長さを表示する（len関数、print関数）
- ファイルを閉じる（close関数）

補足：末尾の改行文字を除かずに塩基配列の長さを調べると、一文字を多く評価されます。

````
# ファイルを開く
f_in = open("sequences20.txt", mode="r")

# 一行ずつ読み込む
for line in f_in:
  line = line.rstrip()  # 末尾の改行文字を取り除く
  print( len(line) )   # 塩基配列の長さを表示する

# ファイルと閉じる
f_in.close()
````

````
# ファイルを開く
f_in = open("sequences20.txt", mode="r")

# 一行ずつ読み込む
for line in f_in:
  line = line.rstrip()  # 末尾の改行文字を取り除く



# ファイルと閉じる
f_in.close()
````

### 解答例1

　条件に一致する塩基配列を数える変数を使う方法です。

````
# ファイルを開く
f_in = open("sequences20.txt", mode="r")

# 条件に一致する塩基配列を数えるカウンター
counter = 0

# 一行ずつ読み込む
for line in f_in:
  line = line.rstrip()  # 末尾の改行文字を取り除く

  count_N = line.count("N")  # Nの個数
  seq_size = len(line)         # 塩基配列の長さ
  rate_N = count_N / seq_size  # N含有率

  # N含有率10%未満のとき、カウンターの値に1足す
  if rate_N < 0.1:
    counter = counter + 1
  
print(counter)  # 最終的なカウンターの値を表示

# ファイルと閉じる
f_in.close()
````

### 解答例2

　条件にあった塩基配列を格納するリストを作成する方法です。

````
# ファイルを開く
f_in = open("sequences20.txt", mode="r")

# 条件に一致した塩基配列を入れるリスト
filtered_seq = []

# 一行ずつ読み込む
for line in f_in:
  line = line.rstrip()  # 末尾の改行文字を取り除く

  count_N = line.count("N")  # Nの個数
  seq_size = len(line)         # 塩基配列の長さ
  rate_N = count_N / seq_size  # N含有率

  # N含有率10%未満のとき、リストに追加する
  if rate_N < 0.1:
    filtered_seq.append(line)
  
print( len(filtered_seq) )  # リストの要素数を表示

# ファイルと閉じる
f_in.close()
````

## 課題5（提出不要）

　それぞれの塩基配列を次のルールにしたがって書き換えてください。すなわち、"ATTGNC" であれば、"TAACNG" に変換してください。

変換ルール
- A → T
- T → A
- C → G
- G → C
- Nはそのまま


ヒント: 文字列のループ処理で、一文字ずつ取り出して、置き換える

````
# ファイルを開く
f_in = open("sequences20.txt", mode="r")

# 一行ずつ読み込む
for line in f_in:
  line = line.rstrip()  # 末尾の改行文字を取り除く



# ファイルと閉じる
f_in.close()
````

### 解答例

````
# ファイルを開く
f_in = open("sequences20.txt", mode="r")

# 置き換え用のディクショナリを作成
replace_nucls = { 'N':'N', 'A':'T', 'T':'A', 'C':'G', 'G':'C' }

# 一行ずつ読み込む
for line in f_in:
  line = line.rstrip()  # 末尾の改行文字を取り除く

  replaced_seq = ""  # 置き換え後の塩基配列用の変数

  # 塩基配列を一文字ずつ取り出す
  for nucl in line:
    # ディクショナリを使って置き換えられた塩基を取得して、
    # 文字列に追加する
    replaced_seq = replaced_seq + replace_nucls[nucl]
  
  # 置き換え後の塩基配列を表示
  print(replaced_seq)


# ファイルと閉じる
f_in.close()
````

## 課題7（提出不要）

　与えられた塩基配列を **逆相補鎖配列** に変換する自作関数 `reverse_complement` を作成してください。

### 解答例

````
def reverse_complement(seq):
  # 塩基置換用のディクショナリ
  replace_nucls = { 'N':'N', 'A':'T', 'T':'A', 'C':'G', 'G':'C' }
  # 塩基置換後の文字列のための変数
  replaced_seq = ""  
  # 塩基配列を一文字ずつ取り出す
  for nucl in seq:
    # ディクショナリから置換後の塩基を取得して、文字列の後ろに追加する
    replaced_seq = replaced_seq + replace_nucls[nucl]

  # 置換後の塩基配列を逆順にする
  rc_seq = replaced_seq[::-1]

  # 戻り値として、逆相補鎖配列を返す
  return rc_seq
````

　次のセルでは、作成したreverse_complement関数を使って、すべての配列を逆相補鎖配列に変換します。

````
# ファイルを開く
f_in = open("sequences20.txt", mode="r")

# 一行ずつ読み込む
for line in f_in:
  line = line.rstrip()  # 末尾の改行文字を取り除く

  # 上のセルで作成したreverse_complement関数を使う
  rc_seq = reverse_complement(line)
  # 返ってきた逆相補鎖配列を表示する
  print(rc_seq)

# ファイルと閉じる
f_in.close()
````

## ワンライナー

　以下は、1行のコードで複数の処理をおこなう方法です。プログラミングの世界では、1行で様々な処理をおこなえるコードのことを、**ワンライナー** と呼んでいます。

　上の課題のいくつかは、for構文内の処理をワンライナーで書くことも可能です。以下では、課題3「N含有率10%未満の塩基配列の個数を数える」をワンライナーで処理しています。

　詳しくは説明しませんが、**内包表記**と呼ばれる書き方を使います。


````
# ファイルを開く
f_in = open("sequences20.txt", mode="r")

# N含有率10%未満の塩基配列の個数を数えるワンライナー
print( sum([1 for line in f_in if line.count("N")/len(line.rstrip())<0.1]) )

# ファイルと閉じる
f_in.close()
````

　一般的に、内包表記は、通常のfor構文よりも高速に処理されます。プログラムの実行速度が上がる反面、複雑な内包表記を使うと、コードの見やすさを損なうといったデメリットもあります。
