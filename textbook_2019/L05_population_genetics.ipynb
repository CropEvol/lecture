#### 解答例

````
# Segregating siteの数を求める関数
def compare_seqs(seq1, seq2):
    S=0 # Segregation siteの初期値
    # seq1とseq2を同時にループ
    for nuc1, nuc2 in zip(seq1, seq2):    
        if nuc1 != nuc2:    # seq1の塩基とseq2の塩基が異なる場合、
            S=S+1                # Segregation siteの値を1増やす
    print(S, "sites are different.")  # 結果を出力
    
# 自作関数compare_seqsを使って、2種間のSegregating siteを求める
compare_seqs(Dioscorea_tokoro, Dioscorea_elephantipes)     # D. tokoro-D. elephantipes
compare_seqs(Dioscorea_tokoro, Dioscorea_sylvatica)            # D. tokoro-D. sylvatica
compare_seqs(Dioscorea_sylvatica, Dioscorea_elephantipes)  # D. sylvatica-D. elephantipes
````

### 系統樹を描画する
　上で得た Segregating site の結果を、それぞれの種間の"距離"とみなして、系統樹として描画してみましょう。draw_tree()という関数の中身については、今の段階で理解する必要ありません。
 
　下記コードのS1,S2,S3の値を書き換えて実行してください。
 
```
S1、S2、S3に先ほど求めたSegregation siteを入力する
- S1:  tokoro vs. elephantipes
- S2:  tokoro vs. sylvatica
- S3:  sylvatica vs. elephantipes
```

````
def draw_tree(S1, S2, S3):
    %matplotlib inline
    import matplotlib.pyplot as plt  #matplotlibというライブラリをインポートする(長い名前なのでコードを書くときはpltとする)
    d = S1 + S2 + S3
    a = d/2 - S1
    b = d/2 - S2
    c = d/2 - S3
    plt.plot([0, -a*(2**(1/2))/2], [0, -a/2])
    plt.plot([0, b*(2**(1/2))/2], [0, -a/2])
    plt.plot([0, 0], [1,c])
    
    #図に文字を載せるコード
    plt.text(-a*(2**(1/2))/2 - 30, -a/2, "sylvatica", fontsize=15)
    plt.text(b*(2**(1/2))/2 + 5, -a/2, "elephantipes", fontsize=15)
    plt.text(5, c, "tokoro", fontsize=15)
    
    #図を見やすくするためのコード
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    plt.xticks([])
    plt.yticks([])
    plt.axes().set_aspect('equal', 'datalim')

# S1、S2、S3に先ほど求めたSegregation site数を入力する
S1=0  # tokoro vs. elephantipes
S2=0  # tokoro vs. sylvatica
S3=0  # sylvatica vs. elephantipes
draw_tree(S1, S2, S3)
````

# 3. $F_{ST}$
　**$F_{ST}$ （遺伝的分化度）** とは、集団間の遺伝的な距離を評価する指標であり、これを用いると、調べた集団間が遺伝的に近いか遠いかを明らかにできます。また、ゲノム領域ごとに $F_{ST}$ を求めることで、ある集団に特異的な遺伝子(ゲノム領域)の存在を明らかにできます。

　$F_{ST}$ は、3つのNucleotide diversity($\pi$) を使って算出できます。
- $\pi_{1}$: 集団1（ここでは青森）の Nucleotide diversity
- $\pi_{2}$: 集団2（ここでは京都）の Nucleotide diversity
- $\pi_{12}$: 青森と京都を合わせた集団の Nucleotide diversity

$$ F_{ST} = \frac{\pi_{12} - (\pi_{1} + \pi_{2}) / 2}{\pi_{12}} $$

　なお、$F_{ST}$ は0~1の値で表され、1に近いほど遺伝的に距離があると考えられます。

````
# Fstを求める関数nei_fstを作成
def nei_fst(ref1, alt1, ref2, alt2):
    # 2つの集団を合わせたリストref12を作成する
    ref12 = []
    for ref_n1, ref_n2 in zip(ref1, ref2):
        ref12.append(ref_n1+ref_n2)
    # 2つの集団を合わせたリストalt12を作成する
    alt12 = []
    for alt_n1, alt_n2 in zip(alt1, alt2):
        alt12.append(alt_n1+alt_n2)
    # 各πを求める
    pi1  = diversity(ref1, alt1)      # 実習3で作った関数を使う
    pi2  = diversity(ref2, alt2)
    pi12 = diversity(ref12, alt12)
    # Fstを計算する
    fst = (pi12-(pi1+pi2)/2)/pi12
    return fst   # 呼び出し元にfstの値を返す

# 関数nei_fstを使って、それぞれの集団間のFstを調べる
fst1 = nei_fst(ref_aomori, alt_aomori, ref_kyoto, alt_kyoto)
fst2 = nei_fst(ref_kyoto, alt_kyoto, ref_kagoshima, alt_kagoshima)
fst3 = nei_fst(ref_kagoshima, alt_kagoshima, ref_aomori, alt_aomori)
print(fst1)
print(fst2)
print(fst3)
````
