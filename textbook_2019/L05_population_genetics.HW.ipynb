### ヒント1: 塩基の長さを調べる

　len関数を使うと、文字列の長さ（文字数）や、リストの長さ（要素数）などを調べることができます。

````
a = len(range(10))
b = len([1,2,1,3,5,2,3])
c = len("GCAAAGGGACTT")

print(a) # range()の中には10個の要素
print(b) # リストの中には7個の要素
print(c) # 文字列の中には12文字
````

### ヒント2: 区画数を調べる

　余剰塩基が出ない場合、割り算の商を調べるだけで区画数を求めることができます。余剰塩基が発生する場合、割り算の余りも調べる必要があります。

　下の例では、18塩基の配列を5塩基ずつ区切った場合の区画数です。

````
x = 18//5
if 18 % 5 > 0:
  x = x + 1

print(x)
````
