## ファイルの読み込み方法

　テキストファイルの基本的な処理方法は次の通りです。
- ファイルを開く（open関数）
- 一行ずつ読み込む（for構文）
  - 末尾の改行文字を取り除く（rstrip関数）
  - 何らかの処理をする; 以下では塩基配列の長さを表示する（len関数、print関数）
- ファイルを閉じる（close関数）

補足：末尾の改行文字を除かずに塩基配列の長さを調べると、一文字を多く評価されます。

````
# ファイルを開く
f_in = open("sequences20.txt", mode="r")

# 一行ずつ読み込む
for line in f_in:
  line = line.rstrip()  # 末尾の改行文字を取り除く
  print( len(line) )   # 塩基配列の長さを表示する

# ファイルと閉じる
f_in.close()
````

#### ヒント: 文字列を逆順に並べ替える

　文字列を逆順に並べ替えるには、文字列の**スライス**と呼ばれる機能を使います。詳細は省きますが、次にように書くと文字列を逆順に並べ替えできます。詳しく知りたい方は、次のようなキーワードをWeb検索してください: `python 文字列 スライス ステップ`

```python
文字列[::-1]
```

````
# 文字列を直接逆順に並べ替える
print("elppa"[::-1])

# 変数に入れた文字列を逆順に並べ替える
word = "remmargorp nohtyp a ydaerla era uoy"
print(word[::-1])
````
