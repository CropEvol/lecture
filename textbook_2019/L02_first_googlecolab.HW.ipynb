## 課題1: プログラムを実行する

　下記のセルのプログラムを実行してください。

````
# DNA配列の長さを調べるプログラム
seq = "ATTGATGCCTGTAATAGCCAC"
seq_length = len(seq)

print('Sequence length:', seq_length)
````

