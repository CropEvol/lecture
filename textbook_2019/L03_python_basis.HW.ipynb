## 課題7
　DNA配列の塩基の構成（塩基配列中にA,T,G,Cそれぞれ何個あるか）を返す関数 `count_nucleotides` を作成してください。  
　また、その関数を使って、リスト `pgk1` の各塩基配列の構成を調べてください。

テキストの関連箇所:
- リスト
- 繰り返し（ループ）構文
- 自作関数

````
pgk1 = [
    "TTCCACACCATCGAGGCTTCTAATCAAGCATATGCCGAAGTGATTGTAGTCTACTTATTTAAGCACA",
    "TTCCACACCATCGAGGCTTCTAATCAAGCATATGCCGAAGTGATTGTAGTCTACTTATTTAAGCACA",
    "TTCCACACCATCGAGGCTTCTAATCAAGCGTATGCCGAAGTGATTGTAGTCTACTTATTTAAACACA",
    "TTCCACACCATCGAGGCTTCTAATCAAGCATATGCCGGAGTGATTGTAGTCTACTTATTTAAACACA",
    "TTCCACACCATCGAGGCTTCTAATCAAGCATATGCCGAAGTGATTGTAGTCTACTTATTTAAACACA",
    "TTCCACACCATCGAGGCTTCTAATCAAGCATATGCCGAAGTGATTGTAGTCTACTTATTTAAACACA",
    "TTCCACACCATCGAAGCTTCTAATCAAGCATAGGCTGAAGTGATTTTAGTCTACTTATTTAAACACA",
    "TTCCACACCATCGAAGCTTCTAATCAAGCATATGCTGAAGTGATTGTAGTCTACTTATTTAAACACA",
]

# 以下にプログラムを追記してください。
````

#### ヒント: 自作関数で複数の値を返す

````
def count_aple(s):
  N_a = s.count("a")
  N_p = s.count("p")
  N_l = s.count("l")
  N_e = s.count("e")
  return N_a, N_p, N_l, N_e

a,p,l,e = count_aple("apple")
print(a, p, l, e)  # a, p, l, eの個数

a,p,l,e = count_aple("appplle")
print(a, p, l, e)  # a, p, l, eの個数
````
