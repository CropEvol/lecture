## 7. データファイルをダウンロードする

### 実習7-1 Colabサーバー上のフォルダ/ファイルの閲覧
　まずは、Colabサーバー上のフォルダやファイルを閲覧してみましょう。

```
手順
1. 左上辺りにある「>」をクリックして、左のサブカラム（左ペイン）を表示させる（すでに左ペインが表示されている場合、この操作は不要）
2. 「ファイル」タブをクリックする
3. 「..」をクリックする
4. 「content」をクリックする
5. 「sample_data」をクリックする
```




　最初は、「..」「sample_data」のフォルダが表示されているはずです。
- 「..」: 上位階層のフォルダ
- 「sample_data」: サンプルデータが入ったフォルダ

　上位階層フォルダをクリックすると、さらに多くのフォルダがサーバー上にあるのがわかります。最初に表示されていた「sample_data」が入ったフォルダは「content」フォルダです。  

　「sample_data」の中には、「README.md」や「anscombe.json」のほか、4つのサンプルファイルが入っています。

### フォルダやファイルの場所について

　Colabサーバー上のフォルダやファイルは、階層構造になっています。ほとんどのパソコンも同じように、階層構造でフォルダやファイルを管理しています。

　さて、Colabサーバー上の「california_housing_test.csv」の所在を見てみましょう。

<img src="https://github.com/CropEvol/lecture/blob/master/textbook_2019/images/file_path.png?raw=true" alt="file_path" height="250px">

- 一番左のスラッシュ `/` は、一番大元のフォルダで、**ルート**（**ルートフォルダ**、**ルートディレクトリ**）と呼ばれています。
- ルートの下に、`content/`（contentフォルダ）があります。
- contentフォルダの下に、`sample_data/`（sample_dataフォルダ）があります。
- sample_dataフォルダの下に、`california_housing_test.csv`があります。


　このように、フォルダやファイルがある場所のことを **パス** といいます。フォルダやファイルの住所のようなものです。

　プログラムにデータファイルを読み込ませる場合、このパスが重要になってきます。データファイルのパスをきちんと指定しないと、プログラムはデータの所在がわからず、読み込めません。

　なお、実習内容は、パスについて気を配らなくても済むように設計しています。

　パスのもう少し知りたい場合、 [こちら](https://github.com/CropEvol/lecture/blob/master/textbook_2019/L02_about_path.md) も参照してください。


### 実習7-2 ファイルのダウンロード
Colabサーバー上のサンプルファイル、sample_dataフォルダの「california_housing_test.csv」をダウンロードしてみましょう。

```
手順
1. 左ペインを開く > 「ファイル」タブ
2. 「sample_data」をクリック
3. 「california_housing_test.csv」を右クリック > 「ダウンロード」
```

### Colabサーバー上のデータの永続性について

　Colabサーバーには、永続的にファイルが残ることはありません。ColabサーバーがOFFと同時に、サーバー上のファイルも削除されます。Colabサーバー上にファイルを保存できないので注意してください。
