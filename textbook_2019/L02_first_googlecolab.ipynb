## 1. Google Colaboratoryについて



### Pythonプログラミング
　Google Colaboratory は、**Python** と呼ばれるプログラムの作成と実行できるオンライン環境です。  

　**プログラム**（または **コード** ）を書くことを、**プログラミング** と言います。実際のプログラミングは、次の手順でおこないます。
1. プログラムを書く
1. プログラムを実行する

　実際にプログラミングしてみましょう。
 

#### 実習
　下の灰色の枠（**セル** と言います）に次の１行（コード）を入力して、左の▶︎マークを押してください。
 
```python
print("Hello, Python")
```

　うまく実行されると、すぐ下に「Hello, Python」と表示されます。もし、うまくいかない場合は、次のことを確認してみてください。
- `print` の綴りは正しいか
- 半角英数の括弧`()`を使っているか
- 半角英数のクオテーション`"`を使っているか

　なお、「Hello, Python」の部分は他の文字に置き換えても構いません。自由に変更できる部分です。
 
```python
# 例
print("I am a Python programmer.")
print("Kyoto University")
```

```python
print(2 + 3)
```

## 2. ノートブックを保存する

　Google ColaboratoryでPythonプログラミングをする際、**ノートブック**と呼ばれるファイルにプログラムを書き込んでいくことになります。このページもノートブックに書かれたものです。以降は、実習を通して、ノートブックの使い方を学んでいきます。

　まずは、ノートブックの保存方法です。また、確認方法も一緒に学びましょう。
  
#### ノートブックの保存
- 画面上部にある「ファイル」 > 「保存」

#### 保存されたノートブックファイルの確認
- 画面上部にある「ファイル」 > 「ドライブで探す」

　保存されたノートブックファイルは、自身のGoogle Drive（クラウドストレージ）に保存されているはずです。

## 3. ノートブックの名前を変更する
　ノートブックの名前（ファイル名でもある）も簡単に変更可能です。

#### ノートブック名の変更
- 画面上部にある「ファイル」 > 「名前の変更...」
- 画面上部に表示されているファイル名を直接クリックする
　

## 4. 以前保存したノートブックを開く

　今度は、Google Driveに保存されているノートブックを開いてみましょう。

#### ノートブックを開く
- 画面上部にある「ファイル」 > 「ノートブックを開く...」

## 5. プログラムを書く/実行する

　このノートブックの最初の節「1. Google Colaboratoryについて」でも簡単なプログラムを書いて、実行しましたが、おさらいしましょう。
 
#### プログラムを書く
1.  画面上部の「+ コード」をクリックして、**コードセル** を作成する（すでにコードセルを作成済みの場合、この操作は不要）
2.  作成したコードセルにプログラムを書く   

#### プログラムを実行する
- コードセルの左にある「▶︎」をクリックする
- コードセルを選択し、「Shift + Enter」キーを押す
- [Windowsの場合] コードセルを選択し、「Ctrl + Enter」キーを押す
- [Macの場合] コードセルを選択し、「control + Enter」キーを押す

#### 実習

　このセルのすぐ下にコードセルを作成して、以下のプログラムを書いてみましょう。「Hello」を10個表示するプログラムです。

```python
for i in range(10):
  print('Hello')
```

_注意1: 半角英数でコードを書いています。_  
_注意2: `for i in range(10):` の行の後ろには、半角の **コロン `:`** があります。_  
_注意3: `print('Hello')` の行には、行頭に半角スペース2個分程度の **字下げ（インデント）** があります。_

## 6. テキストを書く

　コードセルとは別に、**テキストセル** と呼ばれる、プログラムに関する説明文などの文章を書き込むセルもあります。この文章もテキストセルに書いています。
 
#### テキストを書く
1. 画面上部の「+ テキスト」をクリックし、テキストセルを作成する
2.  作成したテキストセルに文章を書くプログラムを書く

#### 実習
　このセルのすぐ下にテキストセルを作成して、以下の文章を書いてみましょう。

```
ゲノム情報解析入門
```

## 7. データファイルをダウンロードする

　ここでは、Google Colaboratoryにあるサンプルデータファイルをダウンロードしてみます。
 
#### サーバー上のフォルダ/ファイルの閲覧
1. 左上辺りにある「>」をクリックして、左のサブカラム（左ペイン）を表示させる（すでに左ペインが表示されている場合、この操作は不要）
1. 左ペインの「ファイル」タブをクリックすると、Google Colaboratoryサーバー上のフォルダやファイルを見れるようになります

　最初は、「..」「sample_data」のフォルダが表示されているはずです。
- 「..」: 上位階層のフォルダ
- 「sample_data」: サンプルデータが入ったフォルダ

　なお、最初に表示されているファイルシステム上の場所（**パス**）は、`/content`です

#### Google Colaboratoryのファイルシステムについて
　こちらを参照してください: [UNIX/Linuxファイルシステム]()
 
#### ファイルのダウンロード
- ファイル名を右クリックして、「ダウンロード」を選ぶ

#### 実習
　左ペインを表示させて、sample_dataフォルダの下にある「california_housing_test.csv」をダウンロードしてください。

## 8. データファイルをアップロードする

　データファイルのアップロードも簡単にできます。
 
#### ファイルのアップロード
- 左ペインの「ファイル」タブ > アップロード > ファイルを選択する
- 任意のフォルダに、データファイルをドラッグ＆ドロップする


## 9. ノートブックをダウンロードする
　**課題提出の際にこの操作が必要です**

　ノートブックを手元のパソコンにダウンロードすることも可能です。
 
　なお、「2. ノートブックを保存する」でおこなった**保存**は、クラウドストレージにファイルを残す操作です。一方、**ダウンロード**は手元のパソコンにファイルを残す操作です。
#### ノートブックのダウンロード
- 画面上部にある「ファイル」 > 「.ipynb をダウンロード」

#### ノートブックのアップロード
　ダウンロードしたノートブックファイル（.ipynb）を再度開きたい場合、アップロードすれば開くことが可能です。
- 画面上部にある「ファイル」 > 「ノートブックをアップロード」
